```python
import matplotlib.pyplot as plt
from astropy.io import fits

# Open the FITS file
file_path = "AS209_edge.fits"
with fits.open(file_path) as hdul:
    # Access the primary HDU (Header Data Unit) data
    image_data = hdul[0].data

# Display the image
plt.imshow(image_data, cmap="gray", origin='lower')
plt.colorbar()  # Add a colorbar to show intensity scale
plt.title("FITS Image")
plt.xlabel("X Pixel")
plt.ylabel("Y Pixel")
plt.show()


# Open the FITS file
file_path = "elipsesimple1.fits"
with fits.open(file_path) as hdul:
    # Access the primary HDU (Header Data Unit) data
    image_data = hdul[0].data

# Display the image
plt.imshow(image_data, cmap="gray", origin='lower')
plt.colorbar()  # Add a colorbar to show intensity scale
plt.title("FITS Image 2")
plt.xlabel("X Pixel")
plt.ylabel("Y Pixel")
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

center = (119, 80)
alpha = 69.731628
beta = 39
theta = -2.751766 * (180/3.14159)

height = 2*alpha
width = 2*beta

fig, ax = plt.subplots()
ax.set(xlim=(0, 255), ylim=(0, 255), aspect="equal")
ellipse = Ellipse(center, height, width, angle=theta)
ax.add_artist(ellipse)
ellipse.set_edgecolor('k')
ellipse.set_facecolor('w')

plt.show()
```
